```python
# A program processes a list of numbers and needs to find the largest value.
def large_num(lst):
    largest = lst[0]
    for i in lst:
        if(i>largest):
            largest = i
            print(i, "is the largest number")
    return largest
```

```python
large_num([65,23,12,56,43])
```

```python
#Simple way
def large(numbers):
    largest = max(numbers)
    return largest
    
print("Largest value:", large([10, 25, 5, 78, 30]))
```

```python
# A company provides employees with a 10% bonus if their salary exceeds $50,000.
def salary(number):
    
    if(number>50000):
        bonus = number / 100 *10
        print("eligible for bonus:", bonus)
        return bonus
    else:
        print("Not eligible for bonus")
        return 0

print("Bonus:", salary(65000))
```

```python
# A text-processing tool reverses a given word or sentence for formatting purposes.
def reversed_text(text):
    reversed_text = text[::-1]
    return reversed_text

print(reversed_text("Hello world python"))
```

```python
# A retail store offers a 20% discount if a customer’s total order exceeds $100.
def retail_store(amount):
    if amount > 100:
        discount = amount * 0.20
        return discount
    else:
        print("No discount")
        return 0

print(retail_store(50), "Discount for your purchase")
```

```python
# A banking system processes withdrawal requests and ensures the user has enough balance.
def bank_balance(amount):
    acct_balance = int(input("Enter the acct balance"))
    if amount < acct_balance:        
        balance = acct_balance - amount
        return balance
    else:
        print("Not eligible to withdrawl balance is low", acct_balance )
        return acct_balance
        
print(bank_balance(3000))
                       
```

```python
def leap_year(year):
    if (year % 4 ==0 and year % 100 != 0) or (year % 400 ==0):
        print("Leap year", year)
        return year
    else:
        print("Not a leap year", year)

leap_year(2000)
```

```python
num = int(input("Enter the number:"))
original = num

num_str = str(num)
digits = len(num_str)
print(digits)

total = 0
for digit in num_str:
    total += int(digit) ** digits

if(total == original):
       print("Armstrong number:", total)
else:
       print("Not an armstrong num")
        
    
```
