```python
# Write a Python function to find the factorial of 5. Expected Output:120
def facto(n):
    for i in range(1,5):
        n = n*i
    print(n)
```

```python
# Write a Python program to check if a number is prime. Expected Output (for input 7):
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True
        
```

```python
is_prime(10)
```

```python
def vowelcount(name):
    vowels = "AEIOUaeiou"
    n=0
    for char in name:
        if char in vowels:
            n += 1
    return n
```

```python
vowelcount("hello")
```

```python
def count_vowels(s):
    return sum(1 for char in s if char.lower() in "aeiou")
```

```python
print("Number of vowels:", count_vowels("Hello"))
```

```python
def fibo(n):
    fib = [0,1]
    for i in range(2,n):
        fib.append(fib[i-1] + fib[i-2])
    return fib
```

```python
fibo(10)
```

```python
def palindrome(text):
    return text == text[::-1]

print(palindrome("hello"))
print(palindrome("madam"))
```
